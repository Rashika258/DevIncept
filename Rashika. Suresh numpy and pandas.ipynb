# Task 3 - Find out the per unit price from the data

```python
#task 3a Convert date time field into pandas time object
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.info()
```

```python
#task 3b Find out the price per unit from the data and create a column for it in the same data
df['Price per unit']=df['Sales']/df['Quantity']-df['Profit']
```

```python
df
```

# Task4 - Find out the monthly revenue and analyze the findings

```python
df1 = pd.DataFrame()
df1['year_month']=df['Order Date'].dt.to_period("M")
df1['monthly_revenue']= df['Sales']*df['Quantity']
monthly_revenue=df1.groupby(df1['year_month'])
monthly_revenue.head()
monthly_revenue=monthly_revenue.sum().reset_index()
monthly_revenue.head()
```

```python
sns.catplot(x='year_month', y='monthly_revenue', kind='bar',data=monthly_revenue)
```

# Task5 - Find out the yearly revenue and analyze the findings

```python
df2 = pd.DataFrame()
df2['year']=df['Order Date'].dt.to_period("Y")
df2['yearly_revenue']= df['Sales']*df['Quantity']
yearly_revenue=df2.groupby(df2['year'])
yearly_revenue=yearly_revenue.sum().reset_index()
yearly_revenue.head()
```

```python
sns.catplot(x='year', y='yearly_revenue', kind='bar',data=yearly_revenue)
```

# Task6 - Finding out the monthly growth rate and analyse the findings

```python
monthly_revenue['growth_rate']= monthly_revenue['monthly_revenue'].diff()
monthly_revenue
```

```python
monthly_revenue['growth_rate'].max()
```

```python
monthly_revenue.plot(x='year_month',y='growth_rate',marker='o')
```
